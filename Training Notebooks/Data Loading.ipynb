```python
import pyedflib
import numpy as np
from scipy.io import wavfile
import xmltodict

def save_audio_to_wav(signal, sample_rate, filename):
    """
    Save an audio signal as a WAV file.

    Parameters:
    - signal: numpy array of audio samples
    - sample_rate: sampling rate of the audio in Hz
    - filename: output filename (should end with .wav)
    """
    # Ensure the signal is in the correct data type
    signal = np.array(signal, dtype=np.float32)

    # Normalize the signal to the range [-1, 1]
    signal = signal / np.max(np.abs(signal))

    # Convert to 16-bit PCM
    signal = (signal * 32767).astype(np.int16)

    # Write the WAV file
    wavfile.write(filename, sample_rate, signal)

audio_signals = []
```

```python
ids = [995, 999, 1000, 1006, 1008, 1010, 1014, 1016, 1018, 1020, 1022, 1024, 1026, 1028, 1037, 1039]
id_str = [f'{x:04d}' for x in ids]

counts_by_type = {
    "Hypopnea" : 0,
    "Snore" : 0,
    "ObstructiveApnea" : 0,
    "MixedApnea" : 0
}

from tqdm import tqdm

for id in tqdm(id_str, "Processing Data"):
    audio_signals = []

    cur_sample_rate = None
    for x in tqdm(range(1, 6), desc=f"Processing Subject {id}"):
        filename = f"./data/RawApneaData/{id}_{x}.edf"
        f = pyedflib.EdfReader(filename)
        signal_labels = f.getSignalLabels()
        signals = {}
        for i, label in enumerate(signal_labels):
            # Read the signal data
            signal = f.readSignal(i)

            # Get additional information about the signal
            sample_rate = f.getSampleFrequency(i)
            physical_dim = f.getPhysicalDimension(i)

            # Store the signal and its metadata in the dictionary
            signals[label] = {
                'data': signal,
                'sample_rate': sample_rate,
                'physical_dimension': physical_dim
            }
            cur_sample_rate = int(sample_rate)
        audio_signals.append(signals["Mic"]["data"])
        f.close()

    # To correct a mistake in file naming...
    audio_signals = [audio_signals[0], audio_signals[1], audio_signals[2], audio_signals[3], audio_signals[4]]
    audio_signal = np.concatenate(audio_signals)

    # Load and parse the RML file
    with open(f'./data/RawApneaData/{id}_CLEAN.rml', 'r') as file:
        rml_content = file.read()
        rml_dict = xmltodict.parse(rml_content)

    events = rml_dict["PatientStudy"]["ScoringData"]["Events"]["Event"]

    event_by_type = {
        "Hypopnea" : [],
        "Snore" : [],
        "ObstructiveApnea" : [],
        "MixedApnea" : []
    }
    for event in events:
        if event["@Type"] in ["Hypopnea", "Snore", "ObstructiveApnea", "MixedApnea"]:
            event_by_type[event["@Type"]].append((float(event["@Start"]), float(event["@Duration"])))

    for key in ["Hypopnea", "Snore", "ObstructiveApnea", "MixedApnea"]:
        if len(event_by_type[key]) == 0:
            continue
        for (st, dur) in event_by_type[key]:
            cur_signal = audio_signal[int(cur_sample_rate*float(st)):int(cur_sample_rate*(float(st)+float(dur)))]
            save_audio_to_wav(cur_signal, cur_sample_rate, f"./data/SoundWAVs/{key}_{counts_by_type[key]}.wav")
            counts_by_type[key] += 1
    print(f"Done with {id}")
```
